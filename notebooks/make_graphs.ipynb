```python
import datetime
import altair as alt
import pickle as pkl
import pandas as pd
from collections import Counter
```

```python
cb_fp = '/Users/matthewdeverna/Documents/Projects/top_fb_domains_trackers/data/2022_08_18/citizen_browser_entities_by_month_2022-08-18.pkl'
top10_fp = '/Users/matthewdeverna/Documents/Projects/top_fb_domains_trackers/data/2022_08_18/citizen_browser_entities_by_month_2022-08-18.pkl'
```

```python
with open(cb_fp, 'rb') as f:
    cb_data = pkl.load(f)
```

```python
cb_counted_data = dict()

for date, entities in cb_data.items():
    cb_counted_data[date] = Counter(entities)
```

```python
frames = []

for date_key, counted in cb_counted_data.items():
    
    temp_frame = pd.DataFrame.from_dict(cb_counted_data[date_key], orient='index').reset_index()
    temp_frame = temp_frame.rename(columns={
        'index':'source',
        0:'num_instances'
    })
    
    pandas_dt = pd.to_datetime(datetime.datetime.strptime(date_key, "%Y_%m"))
    temp_frame['date'] = pandas_dt
    
    frames.append(temp_frame)
    
```

```python
big_frame = pd.concat(frames)
```

```python
totals = big_frame.groupby("source")['num_instances'].sum().reset_index()

totals = totals.sort_values('num_instances', ascending = False)

bars = alt.Chart(totals.head(50)).mark_bar().encode(
    x = alt.X('num_instances:Q', title = 'number of days in top five'),
    y = alt.Y('source:N', sort='-x', title=None)
)

text = alt.Chart(totals.head(50)).mark_text(dx=15, fontSize=15).encode(
    x = alt.X('num_instances:Q'),
    y = alt.Y('source:N', sort='-x'),
    text = alt.Text('num_instances:Q')
)

alt.layer(bars, text).configure_axis(
    titleFontSize=15,
    labelFontSize=14
)
```
